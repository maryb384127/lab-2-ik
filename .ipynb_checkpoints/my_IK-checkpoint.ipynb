# IKpy Quickstart #

Import the IKPy module : 

```python
import ikpy.chain
import numpy as np
import ikpy.utils.plot as plot_utils
import math
import random
from ikpy.utils import geometry
```

The basic element of IKPy is the kinematic `Chain`.
To create a chain from an URDF file : 

```python
my_chain = ikpy.chain.Chain.from_urdf_file("./ur5/ur5_gripper.urdf")
```

# Inverse kinematics

In Inverse Kinematics, you want your kinematic chain to reach a 3D position in space.

To have a more general representation of position, IKPy works with homogeneous coordinates. Homogenous coordinates are represented as a 4x4 matrix storing both position and orientation. Here we only consider (x,y,z) position, not orientation of the chain.

```python
target_position = [ 0, 0, 1]
```

```python
print("The joint angles are : ", my_chain.inverse_kinematics(target_position))
```

You can check that the Inverse Kinematics is correct by comparing with the original position vector : 

```python
real_frame = my_chain.forward_kinematics(my_chain.inverse_kinematics(target_position))
print("Computed position vector : %s, goal position vector : %s" % (real_frame[:3, 3], target_position))
array = np.zeros((100, 3))

for i in range(0,100):
    x = random.gauss(0, 1)
    y = random.gauss(0, 1)
    z = random.gauss(0, 1)
    g = 1/math.sqrt((x*x)+(y*y)+(z*z))
    array [i,0] = x/g
    array [i,1] = y/g
    array [i,2] = z/g
```

# Plotting
And finally plot the result : 

```python
%matplotlib notebook
%matplotlib ipympl
import matplotlib.pyplot as plt
fig, ax = plot_utils.init_3d_figure()
for i in range(0,100):
    point = [array[i,0],array[i,1],array[i,2]]
    actual = my_chain.forward_kinematics(my_chain.inverse_kinematics(point))
    ax.scatter(actual[0],actual[1],actual[2])
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)
x, y = np.meshgrid(x, y)
eq = 0.12 * x + 0.01 * y + 0.25
ax.plot_surface(x, y, eq)
my_chain.plot([0,0,0,0,0,0,0,0], ax, target=target_position)
##my_chain.plot(my_chain.inverse_kinematics(target_position), ax, target=target_position)
plt.show()
```

```python
def intersect(x,y,z):
    target_position = [ x, y, z]
    transformation_matrixes = my_chain.forward_kinematics(my_chain.inverse_kinematics(target_position), full_kinematics=True)
        # Get the nodes and the orientation from the transformation matrix
    for (index, link) in enumerate(my_chain.links):
        node = geometry.from_transformation_matrix(transformation_matrixes[index])
    values = [0,0,0,0,0,0,0,0]
    for(index, link) in enumerate(my_chain.links):
        values[index]= 0.12*(node[0][0])+0.01*(node[0][1])+0.25-(node[0][2])
    neg = (values[0]<0)
    result = False
    for (index, link) in enumerate(my_chain.links):
        if(neg and values[index]>0):
            result = True
    return result
## test cases
print(intersect(0,0,0)) ##False
print(intersect(0,0,1)) ##False
print(intersect(1,1,1)) ##False
```
